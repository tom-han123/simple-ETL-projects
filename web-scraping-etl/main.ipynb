```python
import requests
from bs4 import BeautifulSoup
import html5lib
import pandas as pd
```

```python
#taylor swift song list from wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_songs_by_Taylor_Swift'
response = requests.get(url)
```

```python
data = response.text
soup = BeautifulSoup(data, 'html5lib')
```

```python
dataframe = pd.DataFrame(columns=['Song','Artist', 'Writer', 'Album', 'Year'])
table = soup.find_all('tbody')[2]
```

```python
for row in table.find_all('tr'):
    col1 = row.find_all('th')
    song = col1[0].text.strip().replace("\"",'')
    col2 = row.find_all('td')
    if len(col2) > 0:
        artist = col2[0].text.strip()
        writer = col2[1].text.strip()
        album = col2[2].text.strip()
        year = col2[3].text.strip()
        new_row = pd.DataFrame({'Song':song, 'Artist':artist, 'Writer':writer, 'Album':album, 'Year':year}, index=[0])
        dataframe = pd.concat([new_row, dataframe.loc[:]]).reset_index(drop=True)
dataframe
```

```python
#load songs to csv file
targetfile = 'taylor_swift_songs.csv'
dataframe.to_csv(targetfile)
```
