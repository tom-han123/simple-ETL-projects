```python
import pandas as pd
import glob
```

```python
#price in USD$
iphone14_pro = 999
iphone14_promax = 1099
file = 'exchange_rates.csv'
```

```python
#extract exchange_rate for each country
def extract_from_csv(file):
    dataframe = pd.read_csv(file)
    return dataframe
```

```python
extracted_data = pd.DataFrame( columns = ['Currency', 'Rates'])
dataframe1 = extract_from_csv(file)
df = pd.concat([dataframe1, extracted_data.loc[:]]).reset_index(drop=True)
df
```

```python
#use exchange rate to generate the corresponding iphone14 price for each country
def transform(data):
    df['Iphone-14pro'] = round(df['Rates']*iphone14_pro, 2)
    df['Iphone-14promax'] = round(df['Rates']*iphone14_promax, 2)
    df.drop(['Rates'], axis = 1, inplace=True)
    df.set_index('Currency', inplace=True)
    return df
```

```python
transform_data = transform(df)
transform_data
```

```python
#load data to csv file
target_file = 'iphon14_price.csv'
transform_data.to_csv(target_file)
```
